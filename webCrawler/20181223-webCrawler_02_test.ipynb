## § ① 取出本週熱門新片的「中英文名稱」、「上映日期」、「期待度」、「滿意度」，還有「封面照片網址」
## § ② 用一個你覺得適合的型態存到一個變數內

```python
from bs4 import BeautifulSoup
from operator import itemgetter

movies = []

for p in range(1,3):
    #網址可以改
    url = f'https://movies.yahoo.com.tw/movie_thisweek.html?page={p}'
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "html5lib")
    for i, j, k in zip(soup.find_all(class_='release_movie_name'), soup.find_all(class_='release_movie_time'),soup.find_all(class_='release_foto')):
        d = {}
        d['中文名稱'] = i.find(class_='gabtn').text.replace(' ', '').replace('\n', '')
        d['英文名稱'] = i.find(class_='en').find(class_='gabtn').text.replace(' ', '').replace('\n', '')
        d['期待度'] = i.find(class_='levelbox').span.text
        if i.find(class_='starwithnum'):
            d['滿意度'] = i.find(class_='starwithnum').span['data-num']
        else:
            d['滿意度'] = 0     
        d['上映日期'] = j.text
        d['圖片網址'] = k.find('img')['src']
        d['超連結'] =k.find('a')['href']
        movies.append(d)

        
        
#for pr in range(0,14):
#    print(movies[pr])

#newlist = sorted(movies, key=lambda k: k['期待度'], reverse=True) #排序

#print(newlist[0:5])
```
