# Flow Control

## range()

```python
for i in range(1, 3):
    print(i)
```

## zip()

```python
for i, j in zip(['a', 'b', 'c'], [1, 2, 3]):  # zip 接受一系列可疊代的對象作為參數，將對象中對應的元素打包
    print(i, j)
```

## enumerate()

```python
for i,j in enumerate(['a', 'b', 'c']):  # enumerate 傳回一個次序及值成組tuple的疊代
    print(i, j)
```

## 補充：zip 傳入參數的長度不等，則返回list的長度和參數中長度最短的對象

```python
a = [1, 2, 3]
b = [4, 5, 6]
c = [7, 8, 9, 0]
zip1 = zip(a, b) # zip 接受一系列可疊代的對象作為參數，將對象中對應的元素打包成一個個 tuple（元組），然後返回由這些 tuples 組成的 list（列表）
print(list(zip1))

zip2 = zip(a, c)
print(list(zip2))
```

data：指post導服務器的數據，該方法返回一個包含兩個元素的(filename, headers) 元組，filename 表示保存到本地的路徑，header表示服務器的響應頭
