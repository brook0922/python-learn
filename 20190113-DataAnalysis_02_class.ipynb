# 用 loc, iloc, ix 取得資料
## loc 是通過 行號 進行切片
從索引獲取具有特定標籤的行（或列），可以通過 行號 和 行標籤

```python
# 注意中括號的使用
print(df)
print()
print(df.loc['a', 'one']) # 資料 
print(df.loc['a']['one']) # 資料

print(df.loc['a', ['one']]) # 向量 
print(df.loc['a', ['one', 'two']]) # 向量 
print(df.loc['a', 'one':'two']) # 向量 
print(df.loc['a', :]) # 向量 
print(df.loc['a']) # 向量
print()
print(df.loc[['a'], 'one']) # 向量 
print(df.loc[['a', 'b', 'c', 'd'], 'one']) # 向量 
print(df.loc['a':'d', 'one']) # 向量 
print(df.loc[:, 'one']) # 向量
print()
print(df.loc[['a'], ['one']]) # DataFrame 
print(df.loc[['a'], :]) # DataFrame 
print(df.loc[:, ['one']]) # DataFrame 
print(df.loc[:, :]) # DataFrame
```

## iloc 是通過 行標籤 切片至行標籤為 3 那裏，並且包括 3
獲取索引中特定位置的行（或列），所以它只需要整數也就是 行號
