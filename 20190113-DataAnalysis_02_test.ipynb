# <font color='red'>進階11：請根據給定的資料，進行下列操作：
1. Drop missing observations 刪除缺少的觀察結果
2. Drop rows where all cells in that row is NA 刪除該行中所有單元格為NA的行
3. Create a new column full of missing values 創建一個包含缺失值的新列
4. Fill in missing data with zeros 用零填寫缺失的數據
5. Fill in missing in preTestScore with the mean value of preTestScore 使用preTestScore的平均值填寫preTestScore中的缺失
6. Fill in missing in postTestScore with each sex’s mean value of postTestScore 使用postTestScore的每個性別平均值填寫postTestScore中的缺失
7. Select some raws but ignore the missing data points 選擇一些原始數據但忽略丟失的數據點

```python
raw_data = {'first_name': ['Jason', np.nan, 'Tina', 'Jake', 'Amy'], 
        'last_name': ['Miller', np.nan, 'Ali', 'Milner', 'Cooze'], 
        'age': [42, np.nan, 36, 24, 73], 
        'sex': ['m', np.nan, 'f', 'm', 'f'], 
        'preTestScore': [4, np.nan, np.nan, 2, 3],
        'postTestScore': [25, np.nan, np.nan, 62, 70]}
df = pd.DataFrame(raw_data)
df
```

## <font color='green'> 1.刪除缺少的觀察結果

```python
df_no_missing = df.dropna()
df_no_missing
```

## <font color='green'> 2. 列裡所有欄位都是 na 才丟棄

```python
df_cleaned = df.dropna(how='all')
df_cleaned
```

## <font color='green'> 3.創建一個包含缺失值的新列

```python
df['location'] = np.nan
df
```

## <font color='green'> 4.用零填寫缺失的數據

```python
df.fillna(0)
```

## <font color='green'> 5.使用preTestScore的平均值填寫preTestScore中的缺失

```python
df["preTestScore"].fillna(df["preTestScore"].mean(), inplace=True)  # inplace: boolean, 默认值 False。如果为Ture,在原地填满。
df
```

## <font color='green'> 6.使用postTestScore的每個性別平均值填寫postTestScore中的缺失

```python
df["postTestScore"].fillna(df.groupby("sex")["postTestScore"].transform("mean"), inplace=True)   #將sex欄位化為group 回填平均值
df
```
