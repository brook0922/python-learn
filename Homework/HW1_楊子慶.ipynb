```python
from sklearn import datasets
data = datasets.load_iris()
data
```

```python
import pandas as pd
iris = pd.DataFrame(data['data'], columns=data['feature_names'])
iris
```

```python
iris["target"] = data["target"]
iris
```

```python
import seaborn as sns
sns.pairplot(iris)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X=iris.iloc[:,0:4].values, y=iris["target"].values)
```

```python
predict_y = model.predict(X=iris.iloc[:, 0:4].values)
```

```python
sns.regplot(x=predict_y, y=iris["target"].values, fit_reg=True)
```

```python
model.score(X=iris.iloc[:, 0:4].values, y=iris["target"].values)
```

```python
model.coef_
coefs = pd.DataFrame({"column": iris.columns[0:4], "coef": model.coef_})
coefs
```
