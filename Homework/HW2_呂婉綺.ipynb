## KEEP DATA

```python
dataset = pd.read_csv('hour.csv')
del dataset['instant'] 
del dataset['dteday']
del dataset['casual']
del dataset['cnt'] 

print(dataset.info())  #check null
print('--------------------------------------------------')
print(dataset.describe())
```

## remove outlier

```python
print ("Shape Of The Before Ouliers: ", dataset.shape)
dataset = dataset[abs(dataset["registered"] - np.mean(dataset["registered"]) < 3*np.std(dataset["registered"]))] 
print ("Shape Of The After Ouliers: ", dataset.shape)
```

## plot boxplot

```python
for i, var in enumerate(('season', 'yr', 'mnth', 'hr', 'holiday', 'weekday','workingday',
                         'weathersit')):
    plt.figure(figsize=(10, 4))
    sns.boxplot(x = var , y ='registered', data = dataset)
    plt.show()
```

由圖可知, 季節: 1 =春季，2 =夏季，3 =秋季，4 =冬季，春季的離群值多於其他季節且註冊人數少於其他季節

由圖可知, 年: 0 =2011，1 =2012, 2012年註冊人數比2011年多，2011離群值比2012多

由圖可知, 月: 6月與9月註冊人數是高峰

由圖可知, 時: 在早上7~8點和晚上17~19點時註冊人數會較高

由圖可知, 是否是假期: 1=是, 0=否 非假期時註冊人數比假期註冊人數多

由圖可知, 星期幾: 星期一到星期五的註冊人數比星期六和日多

由圖可知, 是否為工作日:  1=是, 0=否, 工作日的註冊人數比非工作日多

由圖可知, 天氣, 註冊人數:option(1)>option(2)>option(3)>option(4)

```python
for i, var in enumerate(('temp', 'atemp', 'hum', 'windspeed')):
    plt.figure(figsize=(100, 40))
    sns.jointplot(x = var , y ='registered', data = dataset)
    plt.show()
```

由圖可知, 溫度很高或是很低時註冊人數較少

由圖可知, 體感温度很高或是很低時註冊人數較少

由圖可知, 相對濕度很低的時候註冊人數較少

由圖可知, 風速很高的時候註冊人數較少

```python
import numpy as np
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.multiclass import OneVsRestClassifier
from scipy import interp
from sklearn import metrics
```

```python
x = dataset.iloc[:,:11]
y = dataset.iloc[:,11]
#x
#y
# shuffle and split training and test sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=.5,random_state=555)
train = X_train.join(y_train)
test = X_test.join(y_test)
```

```python
svm_rbf = svm.SVC(kernel='rbf', probability=True, gamma='auto', random_state=555)
model = svm_rbf.fit(X_train,y_train)
score = model.score(X_train,np.array(y_train))
print('traning_score=', score)
```

```python
predict_y_train = model.predict(X_train)
accuracy = metrics.accuracy_score(predict_y_train, np.array(y_train))
precision = metrics.precision_score(predict_y_train, np.array(y_train), average='weighted')
recall = metrics.recall_score(predict_y_train, np.array(y_train), average='weighted')
print('traning_accuracy_score =', accuracy )
print('traning_precision_score =', precision )
print('traning_recall_score =', recall )
```

```python
svm_rbf = svm.SVC(kernel='rbf', probability=True, gamma='auto', random_state=555)
model = svm_rbf.fit(X_test,y_test)
score = model.score(X_test,np.array(y_test))
print('test_score=', score)
```

```python
predict_y_test = model.predict(X_test)
accuracy = metrics.accuracy_score(predict_y_test, np.array(y_test))
precision = metrics.precision_score(predict_y_test, np.array(y_test), average='weighted')
recall = metrics.recall_score(predict_y_test, np.array(y_test), average='weighted')
print('traning_accuracy_score =', accuracy )
print('traning_precision_score =', precision )
print('traning_recall_score =', recall )
```
