### IRIS 資料集分類

匯入 pandas 和 seaborn資料集，及由sklearn套件載入datasets模組。


```python
import pandas as pd
from sklearn import datasets
import seaborn as sns
```

載入資料集，並查看內容

```python
data = datasets.load_iris()
data
```

資料結構轉換成DataFrame

```python
flower = pd.DataFrame(data["data"], columns=data["feature_names"])
flower["target"] = data["target"]
flower
```

匯出資料分布的狀態，X軸為sepal length (cm) ,Y軸為speal widht (cm)，target的分類以不同的顏色區分，數據為flower。

```python
sns.relplot(x="sepal length (cm)", y="sepal width (cm)", hue="target", data=flower)
```

從資料集中的X軸全部，Y軸2後的全部，以兩兩一對方式描繪出資料關係圖，

```python
sns.pairplot(hue="target",data=flower.iloc[:,2:])
```

由scikit-learn匯入SVC和線性SVC分類器

```python
from sklearn.svm import SVC, LinearSVC
```

X軸為flower的第2行開始的值，Y軸為target的值

```python
X = flower.iloc[:,0:2].values
y = flower["target"].values
```

使用線性SVC模型，亂數設為555

```python
model = LinearSVC(random_state=555)
```

帶入X及y到線性SVC的演算法訓練

```python
model.fit(X=X, y=y)
```

訓練後的分數為0.8

```python
model.score(X=X, y=y)
```

數據預測

```python
predict_y = model.predict(X=X)
```

```python
由sklearn套件載入metrics模組。
```

```python
from sklearn import metrics
```

```python
混淆矩陣帶入y及預測的y
```

```python
confmat = metrics.confusion_matrix(y, predict_y)
```

```python
以Seaborn熱力圖顯示，顏色為YLGBu，格子內填入數據值，格式設置為數字
```

```python
sns.heatmap(confmat, cmap="YlGnBu_r", annot=True, fmt="d")
```

```python
分類結果報告
```

```python
report = metrics.classification_report(y, predict_y)
print(report)
```

```python
分類準確性分數
```

```python
metrics.accuracy_score(y, predict_y)
```
