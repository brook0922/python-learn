## I. 建立預測或是分類模型

本回作業採用Python的Sklearn機器學習庫中的資料集-IRIS資料集。用以分析資料集之間特徵的關係，根據花瓣長度、花瓣寬度、花萼長度、花萼寬度四個特徵進行花朵類型的分類 => 建立分類模型。

### 將 data 轉成 dataframe - 二維格式,以方便觀察資料

```python
iris = pd.DataFrame(data["data"], columns=data["feature_names"])
iris["target"] = data["target"]
iris
```

###  確認資料的完整性

```python
iris.notnull().all() 
```

# 檔案資料完整,沒有需要填補的資料

### 由於 iris 資料庫是屬於分類性質的資料集是要處理因變數為分類變數的迴歸問題
### 屬於一種分類方法,所以應運用邏輯迴歸（Logistic Regression）來解析此題
### 接著透過 seaborn pairplot 觀察 iris 各column 彼此相互關聯 

```python
sns.pairplot(hue="target", data=iris.iloc[:, :])
```

###  依上列圖形觀察確實有分群的趨勢,所以進行SVM model training

```python
from sklearn.svm import SVC, LinearSVC
X = iris.iloc[:, 0:4].values
y = iris["target"].values
model = LinearSVC(random_state=555)
model.fit(X=X, y=y)
```

###  Mean accuracy 查看 model 準確率的評分

```python
model.score(X=X, y=y)
```

## 分數 0.9666666666666667 非常好
### 接著 以 confusion matrix 來分析 

```python
#  Confusion matrix

from sklearn import metrics
predict_y = model.predict(X=X)
confmat = metrics.confusion_matrix(y, predict_y)
confmat
```

```python
sns.heatmap(confmat, cmap="YlGnBu_r", annot=True, fmt="d")
```

## 從Confusion matrix混淆矩陣來看
####  class=0 50/50 全數正確
####  class=1 47/50 有 3 個分錯
####  class=2 48/50 有 2 個分錯

### 最終來看分類總結報告  Classification report

```python
report = metrics.classification_report(y, predict_y)
print(report)
```

####  class=0  precision & recall  &  f1-score 全數=1 ,表示正確完美
####  class=1  precision 預測準確率=0.96  recall  回測正確率=0.94 總和f1-score=0.95
####  class=2  precision 預測準確率=0.94  recall  回測正確率=0.96 總和f1-score=0.95

## 整體而言分數都很好.
